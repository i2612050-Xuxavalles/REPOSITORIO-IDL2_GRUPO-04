**Ejercicio 1: Calcular el total que una persona debe pagar en una llantera, si el precio de cada llanta es de S/150 si se compran menos de 5 lantas y de S/120 si se compran 5 o más.**

```python
cantidad_llantas = int(input("Ingrese la cantidad de llantas compradas: "))

if cantidad_llantas < 5:
    precio_llanta = 150
else:
    precio_llanta = 120

total_pagar = cantidad_llantas * precio_llanta

print("Precio por llanta: S/", precio_llanta)
print("Total a pagar: S/", total_pagar)
```

**Ejercicio 2: Cómo se podría determinar ¿Cuál de los 3 números brindados por usuario es el menor y mayor de todos ellos?**

```python
numero1 = float(input("Ingrese el primer número: "))
numero2 = float(input("Ingrese el segundo número: "))
numero3 = float(input("Ingrese el tercer número: "))

# Determinar el mayor
if numero1 >= numero2 and numero1 >= numero3:
    mayor = numero1
elif numero2 >= numero1 and numero2 >= numero3:
    mayor = numero2
else:
    mayor = numero3

# Determinar el menor
if numero1 <= numero2 and numero1 <= numero3:
    menor = numero1
elif numero2 <= numero1 and numero2 <= numero3:
    menor = numero2
else:
    menor = numero3

print("El número mayor es:", mayor)
print("El número menor es:", menor)
```

**Ejercicio 4: En una escuela la colegiatura de los alumnos se determina según el número de materias que cursan (Solicitar al usuario). El costo de todas las materias es el mismo S/100. Se ha establecido un programa para estimular a los alumnos, el cual consiste en lo siguiente: si el
promedio obtenido por un alumno en el último periodo es mayor o igual que 18, se le hará un descuento del 50% sobre la mitad de los cursos; si el promedio obtenido es menor que 18 deberá pagar la colegiatura completa. Obtener cuanto debe pagar un alumno.**

```python
numero_materias = int(input("Ingrese el número de materias que cursa: "))
promedio = float(input("Ingrese el promedio obtenido en el último periodo: "))

costo_materia = 100
colegiatura = numero_materias * costo_materia

if promedio >= 18:
    descuento = (colegiatura / 2) * 0.50
    total_pagar = colegiatura - descuento
else:
    descuento = 0
    total_pagar = colegiatura

print("Colegiatura sin descuento: S/", colegiatura)
print("Descuento aplicado: S/", descuento)
print("Total a pagar: S/", total_pagar)
```

**Ejercicio 5: Determinar mediante el ingreso de los 3 lados de un triángulo, si este es Equilátero (Mostrar el perímetro del), Isósceles (Indicar que lados son Iguales) o Escaleno (triángulo mostrar el área del triángulo).**

```python
lado1 = float(input("Ingrese el primer lado del triángulo: "))
lado2 = float(input("Ingrese el segundo lado del triángulo: "))
lado3 = float(input("Ingrese el tercer lado del triángulo: "))

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:

    if lado1 == lado2 and lado2 == lado3:
        perimetro = lado1 + lado2 + lado3
        print("El triángulo es equilátero.")
        print("El perímetro del triángulo es:", perimetro)

    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("El triángulo es isósceles.")

        if lado1 == lado2:
            print("Los lados iguales son lado 1 y lado 2.")
        elif lado1 == lado3:
            print("Los lados iguales son lado 1 y lado 3.")
        else:
            print("Los lados iguales son lado 2 y lado 3.")

    else:
        semiperimetro = (lado1 + lado2 + lado3) / 2
        area = (semiperimetro *
                (semiperimetro - lado1) *
                (semiperimetro - lado2) *
                (semiperimetro - lado3)) ** 0.5

        print("El triángulo es escaleno.")
        print("El área del triángulo es:", area)

else:
    print("Los lados ingresados no forman un triángulo.")
```

**Ejercicio 6: Elabore un programa que lea la hora (HH MM SS) y muestre por pantalla la hora N segundos después.**

```python
hora = int(input("Ingrese la hora: "))
minuto = int(input("Ingrese los minutos: "))
segundo = int(input("Ingrese los segundos: "))
n_segundos = int(input("Ingrese la cantidad de segundos a aumentar: "))

if hora >= 0 and hora < 24 and minuto >= 0 and minuto < 60 and segundo >= 0 and segundo < 60:

    total_segundos = hora * 3600 + minuto * 60 + segundo
    total_segundos = total_segundos + n_segundos

    nueva_hora = (total_segundos // 3600) % 24
    resto = total_segundos % 3600
    nuevo_minuto = resto // 60
    nuevo_segundo = resto % 60

    print("La nueva hora es:")
    print(nueva_hora, ":", nuevo_minuto, ":", nuevo_segundo)

else:
    print("La hora ingresada no es válida.")
```

**Ejercicio 7: Calcular el pago de una llamada, el cobro es por el tiempo que ésta dura, de tal forma que los primeros cinco minutos cuestan S/ 1.00 c/u, los siguientes tres, S/0.80 c/u, los siguientes dos minutos, S/0.70 c/u, y a partir del décimo minuto, S/0.50 c/u. Además, se carga un impuesto de 3 % cuando es domingo, y 5% si es domingo de turno nocturno. Realice un algoritmo para determinar cuánto debe pagar por cada concepto una persona que realiza una llamada.**

```python
minutos = int(input("Ingrese la cantidad de minutos de la llamada: "))
dia = input("Ingrese el día de la llamada: ")
turno = input("Ingrese el turno de la llamada (diurno/nocturno): ")

if minutos <= 5:
    costo_llamada = minutos * 1.00

elif minutos <= 8:
    costo_llamada = 5 * 1.00 + (minutos - 5) * 0.80

elif minutos <= 10:
    costo_llamada = 5 * 1.00 + 3 * 0.80 + (minutos - 8) * 0.70

else:
    costo_llamada = 5 * 1.00 + 3 * 0.80 + 2 * 0.70 + (minutos - 10) * 0.50

if dia == "domingo" and turno == "nocturno":
    impuesto = costo_llamada * 0.05
elif dia == "domingo":
    impuesto = costo_llamada * 0.03
else:
    impuesto = 0

total_pagar = costo_llamada + impuesto

print("Costo de la llamada: S/", costo_llamada)
print("Impuesto aplicado: S/", impuesto)
print("Total a pagar: S/", total_pagar)
```

**Ejercicio 8: Solicite a usuario 3 números enteros (validar que sean diferentes) y en que orden lo desean ver, para luego, mostrarlos a usuario, pero ordenados de manera ascendente o descendente según los que indique el usuario.**

```python
numero1 = int(input("Ingrese el primer número entero: "))
numero2 = int(input("Ingrese el segundo número entero: "))
numero3 = int(input("Ingrese el tercer número entero: "))

orden = input("Ingrese el orden deseado: ascendente o descendente: ")

if numero1 != numero2 and numero1 != numero3 and numero2 != numero3:

    if numero1 < numero2 and numero1 < numero3:
        menor = numero1
        if numero2 < numero3:
            medio = numero2
            mayor = numero3
        else:
            medio = numero3
            mayor = numero2

    elif numero2 < numero1 and numero2 < numero3:
        menor = numero2
        if numero1 < numero3:
            medio = numero1
            mayor = numero3
        else:
            medio = numero3
            mayor = numero1

    else:
        menor = numero3
        if numero1 < numero2:
            medio = numero1
            mayor = numero2
        else:
            medio = numero2
            mayor = numero1

    if orden == "ascendente":
        print("Orden ascendente:", menor, medio, mayor)
    elif orden == "descendente":
        print("Orden descendente:", mayor, medio, menor)
    else:
        print("Orden no válido.")

else:
    print("Los números deben ser diferentes.")
```

**Ejercicio 9: Permita calcular el total a pagar por la compra de N camisas. Si se compran entre 1 a 4 camisas se aplica un descuento del 12.5%, si se compra una cantidad comprendida entre 5 y 8 camisas se aplica un descuento del 20% y si se compran cantidades mayores, se aplica un
descuento del 31.5% sobre el total de la compra. Debe imprimirse en pantalla la compra final sin descuento, monto del descuento y la compra con descuento.**

```python
cantidad_camisas = int(input("Ingrese la cantidad de camisas compradas: "))
precio_camisa = float(input("Ingrese el precio de cada camisa: "))

compra_sin_descuento = cantidad_camisas * precio_camisa

if cantidad_camisas >= 1 and cantidad_camisas <= 4:
    descuento = compra_sin_descuento * 0.125

elif cantidad_camisas >= 5 and cantidad_camisas <= 8:
    descuento = compra_sin_descuento * 0.20

elif cantidad_camisas > 8:
    descuento = compra_sin_descuento * 0.315

else:
    descuento = 0
    print("Cantidad de camisas no válida.")

compra_con_descuento = compra_sin_descuento - descuento

print("Compra sin descuento: S/", compra_sin_descuento)
print("Monto del descuento: S/", descuento)
print("Compra con descuento: S/", compra_con_descuento)
```
