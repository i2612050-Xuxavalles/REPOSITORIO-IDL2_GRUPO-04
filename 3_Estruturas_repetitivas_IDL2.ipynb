**Ejercicio 1: Calcular la suma y promedio de los números pares e impares por separado, desde 1 hasta N.**

```python
N = int(input("Ingrese el valor de N: "))

suma_pares = 0
suma_impares = 0
cantidad_pares = 0
cantidad_impares = 0

for numero in range(1, N + 1):
    if numero % 2 == 0:
        suma_pares = suma_pares + numero
        cantidad_pares = cantidad_pares + 1
    else:
        suma_impares = suma_impares + numero
        cantidad_impares = cantidad_impares + 1

if cantidad_pares > 0:
    promedio_pares = suma_pares / cantidad_pares
else:
    promedio_pares = 0

if cantidad_impares > 0:
    promedio_impares = suma_impares / cantidad_impares
else:
    promedio_impares = 0

print("Suma de números pares:", suma_pares)
print("Promedio de números pares:", promedio_pares)
print("Suma de números impares:", suma_impares)
print("Promedio de números impares:", promedio_impares)
```

**Ejercicio 2: Realizar un algoritmo que permita pedir 10 números y determine e imprima cuántos son pares, impares, positivos, neutros y negativos.**

```python
pares = 0
impares = 0
positivos = 0
neutros = 0
negativos = 0

for i in range(1, 11):
    numero = int(input("Ingrese el número " + str(i) + ": "))

    if numero % 2 == 0:
        pares = pares + 1
    else:
        impares = impares + 1

    if numero > 0:
        positivos = positivos + 1
    elif numero < 0:
        negativos = negativos + 1
    else:
        neutros = neutros + 1

print("Cantidad de números pares:", pares)
print("Cantidad de números impares:", impares)
print("Cantidad de números positivos:", positivos)
print("Cantidad de números neutros:", neutros)
print("Cantidad de números negativos:", negativos)
```

**Ejercicio 3: Hacer un programa que permita ingresar un conjunto de N números y un número más X. Mostrar como resultado:**
- Porcentaje de números menores que X.
- Porcentaje de números mayores que X.
- Porcentaje de números iguales a X.
- Porcentaje de números comprendidos entre 1 y X.
- Porcentaje de números que no cumplen ninguna condición.

```python
# Ejercicio 3: Porcentajes según comparación con X

N = int(input("Ingrese la cantidad de números: "))
X = int(input("Ingrese el valor de X: "))

menores = 0
mayores = 0
iguales = 0
entre_1_y_x = 0
ninguna = 0

for i in range(1, N + 1):
    numero = int(input("Ingrese el número " + str(i) + ": "))

    if numero < X:
        menores = menores + 1

    if numero > X:
        mayores = mayores + 1

    if numero == X:
        iguales = iguales + 1

    if numero >= 1 and numero <= X:
        entre_1_y_x = entre_1_y_x + 1
    elif numero < 1:
        ninguna = ninguna + 1

porcentaje_menores = (menores / N) * 100
porcentaje_mayores = (mayores / N) * 100
porcentaje_iguales = (iguales / N) * 100
porcentaje_entre_1_y_x = (entre_1_y_x / N) * 100
porcentaje_ninguna = (ninguna / N) * 100

print("Porcentaje de números menores que X:", porcentaje_menores, "%")
print("Porcentaje de números mayores que X:", porcentaje_mayores, "%")
print("Porcentaje de números iguales a X:", porcentaje_iguales, "%")
print("Porcentaje de números comprendidos entre 1 y X:", porcentaje_entre_1_y_x, "%")
print("Porcentaje de números que no cumplen ninguna condición:", porcentaje_ninguna, "%")
```

**Ejercicio 5: Realizar un algoritmo para determinar cuánto ahorrará una persona en un año, si al final de cada mes deposita cantidades variables de dinero. Se quiere saber cuánto lleva ahorrado cada vez que haga un depósito en cada mes y luego mostrar el ahorro final de 12 meses.**

**Ejercicio 6: Hacer un programa que muestre un cronómetro en formato hh:mm:ss, indicando las horas, minutos y segundos. Considerar que el tiempo no será exacto al de un cronómetro real, solo será referencial.**

```python
horas_maximas = int(input("Ingrese la cantidad de horas a mostrar en el cronómetro: "))

for hora in range(0, horas_maximas + 1):
    for minuto in range(0, 60):
        for segundo in range(0, 60):
            print(hora, ":", minuto, ":", segundo)

            if hora == horas_maximas and minuto == 0 and segundo == 10:
                break

        if hora == horas_maximas and minuto == 0:
            break

    if hora == horas_maximas:
        break
```

**Ejercicio 7: Crear una aplicación que pida dos números y calcule el factorial de cada número. El programa también debe indicar hasta qué valor serían iguales.**
Ejemplo:
7! > 1x2x3x4x5x6x7
10! > 1x2x3x4x5x6x7x8x9x10
Son iguales hasta el múltiplo 7.

```python
numero1 = int(input("Ingrese el primer número: "))
numero2 = int(input("Ingrese el segundo número: "))

factorial1 = 1
factorial2 = 1

serie1 = ""
serie2 = ""

for i in range(1, numero1 + 1):
    factorial1 = factorial1 * i

    if i == 1:
        serie1 = str(i)
    else:
        serie1 = serie1 + "x" + str(i)

for i in range(1, numero2 + 1):
    factorial2 = factorial2 * i

    if i == 1:
        serie2 = str(i)
    else:
        serie2 = serie2 + "x" + str(i)

if numero1 < numero2:
    igual_hasta = numero1
else:
    igual_hasta = numero2

print(str(numero1) + "! >", serie1)
print("Resultado:", factorial1)

print(str(numero2) + "! >", serie2)
print("Resultado:", factorial2)

print("Son iguales hasta el múltiplo", igual_hasta)
```

**Ejercicio 8: A una fiesta asistieron personas de diferentes edades y sexos. Construir un algoritmo donde se ingresen los datos de los N asistentes.**

Calcular:
- Cuántas personas asistieron a la fiesta.
- Cuántos hombres y cuántas mujeres.
- Promedio de edades por sexo.
- La edad de la persona de menor y mayor edad que asistió.
Nota: El sistema debe permitir al menos el ingreso de los datos de una persona y solicitar si se desea registrar a otra.

```python
total_personas = 0
total_hombres = 0
total_mujeres = 0

suma_edad_hombres = 0
suma_edad_mujeres = 0

menor_edad = 0
mayor_edad = 0

continuar = "si"

while continuar == "si":
    edad = int(input("Ingrese la edad de la persona: "))
    sexo = input("Ingrese el sexo de la persona (M/F): ")

    total_personas = total_personas + 1

    if total_personas == 1:
        menor_edad = edad
        mayor_edad = edad
    else:
        if edad < menor_edad:
            menor_edad = edad

        if edad > mayor_edad:
            mayor_edad = edad

    if sexo == "M" or sexo == "m":
        total_hombres = total_hombres + 1
        suma_edad_hombres = suma_edad_hombres + edad

    elif sexo == "F" or sexo == "f":
        total_mujeres = total_mujeres + 1
        suma_edad_mujeres = suma_edad_mujeres + edad

    else:
        print("Sexo no válido. No se contará como hombre ni mujer.")

    continuar = input("¿Desea registrar otra persona? Escriba si o no: ")

if total_hombres > 0:
    promedio_hombres = suma_edad_hombres / total_hombres
else:
    promedio_hombres = 0

if total_mujeres > 0:
    promedio_mujeres = suma_edad_mujeres / total_mujeres
else:
    promedio_mujeres = 0

print("Cantidad total de personas:", total_personas)
print("Cantidad de hombres:", total_hombres)
print("Cantidad de mujeres:", total_mujeres)
print("Promedio de edad de hombres:", promedio_hombres)
print("Promedio de edad de mujeres:", promedio_mujeres)
print("Menor edad registrada:", menor_edad)
print("Mayor edad registrada:", mayor_edad)
```

# Ejercicio 9: Calcular el producto utilizando arrays:
100 x 98 x 96 x 94 x ... x 1, en ese orden.

```python
numeros = []

for numero in range(100, 0, -2):
    numeros.append(numero)

producto = 1

for numero in numeros:
    producto = producto * numero

print("Array generado:")
print(numeros)

print("Producto total:")
print(producto)
```

**Ejercicio 10: Desarrollar un programa que para N estudiantes, con un mínimo de 5 y máximo de 20, permita leer sus notas obtenidas en un curso. Se debe validar que la nota tenga un valor entre 0 y 20. La mínima nota aprobatoria es 13. Al final calcular:**
a) La mayor nota.
b) La menor nota.
c) La cantidad de alumnos que obtuvieron la mayor nota.
d) La cantidad de alumnos que obtuvieron la menor nota.
e) El promedio de notas del salón.
f) La cantidad de notas pares.
g) La cantidad de notas impares.
h) La cantidad de alumnos aprobados.
i) La cantidad de alumnos desaprobados.
j) El porcentaje de alumnos aprobados.
k) El porcentaje de alumnos desaprobados.

```python
N = int(input("Ingrese la cantidad de estudiantes, mínimo 5 y máximo 20: "))

while N < 5 or N > 20:
    print("Cantidad no válida.")
    N = int(input("Ingrese nuevamente la cantidad de estudiantes, mínimo 5 y máximo 20: "))

notas = []

for i in range(1, N + 1):
    nota = int(input("Ingrese la nota del estudiante " + str(i) + ": "))

    while nota < 0 or nota > 20:
        print("Nota no válida. Debe estar entre 0 y 20.")
        nota = int(input("Ingrese nuevamente la nota del estudiante " + str(i) + ": "))

    notas.append(nota)

mayor_nota = notas[0]
menor_nota = notas[0]
suma_notas = 0
notas_pares = 0
notas_impares = 0
aprobados = 0
desaprobados = 0

for nota in notas:
    suma_notas = suma_notas + nota

    if nota > mayor_nota:
        mayor_nota = nota

    if nota < menor_nota:
        menor_nota = nota

    if nota % 2 == 0:
        notas_pares = notas_pares + 1
    else:
        notas_impares = notas_impares + 1

    if nota >= 13:
        aprobados = aprobados + 1
    else:
        desaprobados = desaprobados + 1

cantidad_mayor_nota = 0
cantidad_menor_nota = 0

for nota in notas:
    if nota == mayor_nota:
        cantidad_mayor_nota = cantidad_mayor_nota + 1

    if nota == menor_nota:
        cantidad_menor_nota = cantidad_menor_nota + 1

promedio = suma_notas / N
porcentaje_aprobados = (aprobados / N) * 100
porcentaje_desaprobados = (desaprobados / N) * 100

print("Notas ingresadas:", notas)
print("Mayor nota:", mayor_nota)
print("Menor nota:", menor_nota)
print("Cantidad de alumnos con la mayor nota:", cantidad_mayor_nota)
print("Cantidad de alumnos con la menor nota:", cantidad_menor_nota)
print("Promedio de notas del salón:", promedio)
print("Cantidad de notas pares:", notas_pares)
print("Cantidad de notas impares:", notas_impares)
print("Cantidad de alumnos aprobados:", aprobados)
print("Cantidad de alumnos desaprobados:", desaprobados)
print("Porcentaje de alumnos aprobados:", porcentaje_aprobados, "%")
print("Porcentaje de alumnos desaprobados:", porcentaje_desaprobados, "%")
```
